```python
## 1 . drop the unnecessary columns 
# ['EmployeeCount','Over18','StandardHours'] these are those columns in which all values are same
data.drop(columns=['EmployeeCount','Over18','StandardHours'],inplace=True)
```

```python
data.columns
```

```python
# 2. Giving the columns new names (if needed)
# Example: Let's rename some columns for better clarity
new_names_mapping = {
    'BusinessTravel': 'Business_Travel',
    'EducationField' : 'Education_Field',
    'JobRole' :'JobRole',
    'JobSatisfaction':'JobSatisfaction'}
    
data = data.rename(columns=new_names_mapping)
```

```python
## 3.Drop the duplicates values
data = data.drop_duplicates()
```

```python
# 4 drop the Nan value
data = data.dropna()
```

```python
# 5. Sanitizing specific columns
for column in data.columns:
    # Check if the column is of string type
    if data[column].dtype == 'object':
        # Remove leading and trailing whitespaces from column values
        data[column] = data[column].str.strip()
```

```python
# Display the cleaned data
data.head(4)
```

#### Visualization

```python
# 1. Histogram of Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=30, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Bar plot of Attrition
plt.figure(figsize=(8, 6))
sns.countplot(x='Attrition', data=data, palette='Set2')
plt.title('Attrition Count')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Box plot of Monthly Income by Department
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='MonthlyIncome', data=data)
plt.title('Monthly Income by Department')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.show()
```

```python
# Visualize all columns
for column in data.columns:
    if data[column].dtype == 'object':
        # Categorical column: create count plot
        plt.figure(figsize=(8, 6))
        sns.countplot(x=column, data=data)
        plt.title(f'{column} Count')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.show()
    else:
        # Numerical column: create histogram
        plt.figure(figsize=(8, 6))
        sns.histplot(data[column], bins=30, kde=True, color='red')
        plt.title(f'{column} Distribution')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
```
